# Introduction to Asset-Liability Management (ALM)

The main idea of Asset-Liability Management (ALM) is that portfolio decisions must consider not just the assets, but also the **liabilities** they are meant to cover. Focusing only on asset value can be misleading and dangerous.

---




## The Pension Fund Crisis: A Perfect Storm

At the turn of the millennium (1999-2003), a major crisis hit the defined benefit pension plans of S&P 500 companies in the US.

* **The Shock**: In just a few years, these plans went from a collective surplus of **$240 billion** to a collective deficit of over **$250 billion**. A nearly $500 billion swing.
* **This happened again in 2008** during the global financial crisis, leading to severe underfunding across US state pension plans and globally.

### What Caused the Crisis?

This was described as a **"perfect storm"** of two simultaneous, adverse market events:

1.  **Falling Asset Values **:
    * The **tech bubble burst** in 2000-2002, causing a sharp decrease in equity markets.
    * This significantly reduced the value of pension plan assets that were invested in stocks.

2.  **Rising Liability Values **:
    * At the same time, **interest rates were decreasing**.
    * Pension liabilities are calculated as a present value of future payments, much like a bond portfolio. When discount rates (interest rates) go down, the present value of future liabilities goes **up**.

This combination—assets falling while liabilities were rising—was disastrous and exposed major weaknesses in the risk management practices of the time.

---

## The Key Takeaway: What Really Matters in ALM

The central lesson is that the absolute value of assets is not the most important metric. What matters is the **value of assets relative to liabilities**.

* **Good News Scenario**: If your assets go down by 5% but your liabilities go down by 10%, your financial health has actually **improved**.
* **Bad News Scenario**: If your assets go up by 5% but your liabilities go up by 10%, your financial health has **worsened**.

### Key Metrics to Focus On

Instead of just looking at wealth, ALM focuses on two related indicators:

1.  **The Surplus (or Deficit)**:
    * **Formula**: `Surplus = Assets - Liabilities`
    * This is the dollar amount by which assets exceed (or fall short of) liabilities.

2.  **The Funding Ratio**:
    * **Formula**: `Funding Ratio = Assets / Liabilities`
    * This is the most critical indicator. It tells you what percentage of your promised liabilities you can cover with your current assets.
    * A funding ratio of **100% or above** means the plan is fully funded and in good shape.
    * A funding ratio **below 100%** indicates a deficit, meaning the plan does not have enough assets to meet its future obligations without additional contributions.

### Conclusion

A sole focus on growing asset wealth is misleading and incomplete. Effective risk management requires focusing on the funding ratio and ensuring that assets and liabilities move in a coordinated way to protect the plan's surplus.

# Liability Hedging Portfolios (LHP)

introduces Liability Hedging Portfolios (LHP) as a critical building block for any investment strategy that needs to account for future liabilities or goals.

---

## The Core Problem & The LHP Solution

For an investor with liabilities (like a pension fund) or long-term goals (like an individual saving for retirement), the main concern isn't just a drop in asset value. The real risk is an **unexpected increase in the present value of their liabilities relative to their assets**, which causes the funding ratio to fall.

### The Solution: Liability Hedging Portfolios (LHP)

A **Liability Hedging Portfolio (LHP)** is an asset portfolio specifically designed to mimic the behavior of the liabilities.

* **For Individuals**: The equivalent is a **Goal Hedging Portfolio (GHP)**.
* **Definition**: An LHP is a portfolio that generates cashflows which perfectly **match the date and amount** of the liability payments.
* **How it's Built**: Typically constructed using fixed-income instruments like **bonds** (nominal or inflation-linked) or derivatives like **interest rate swaps**.

---

## Example: The "Retirement Bond"

Consider an individual who plans to retire in 2035 and will need a steady income stream for 20 years after that.

* **A Standard Bond is NOT Safe**: A regular bond paying coupons now and a large principal at a single maturity date does not match the investor's needs. The cashflows are misaligned with the goal.
* **A "Retirement Bond" (GHP) is the Safe Asset**: The ideal asset is a custom portfolio that pays **no cashflows until 2035**, and *then* begins paying a series of regular cashflows for the next 20 years. This perfectly matches the retirement income goal.

---

## Redefining the "Safe Asset": The ALM Perspective

How we perceive risk changes completely when we consider liabilities.

* **Asset-Only Perspective (Misleading)**: From this view, a long-duration "Retirement Bond" looks extremely volatile and risky compared to cash. Cash appears to be the safe asset.
* **Funding Ratio Perspective (Correct)**: This is the correct view.
    * The **Retirement Bond** is the true safe asset because its value moves perfectly in sync with the liability value. It has **zero risk** with respect to the funding ratio.
    * **Cash** is now seen as **extremely risky** because its value is stable while the value of the liabilities (the denominator in the funding ratio) fluctuates wildly with interest rates.

> **Key Takeaway**: A "safe asset" cannot be defined in absolute terms. It can only be defined as the asset that is safe **with respect to the investor's specific liability or goal.**

---

## Practical Implementation of LHPs

Perfectly matching liabilities is the goal, but can be hard in practice. There are two main methods:

1.  **Cashflow Matching (Ideal)**
    * Buy zero-coupon bonds with maturity dates and face values that perfectly match each liability payment.
    * This is often **impractical** due to a limited supply of bonds with the exact maturities and amounts needed.

2.  **Duration Matching (Practical Approximation)**
    * When perfect matching isn't possible, the next best thing is to build an asset portfolio that has the **exact same duration** as the liability portfolio.
    * **How it Works**: If interest rates change, the asset portfolio's value will change by approximately the same amount as the liability portfolio's value, thus "immunizing" the surplus against interest rate risk.
    * **Further Refinements**: For better hedging, one can also match **convexity** (a second-order effect) and immunize against changes in the **shape of the yield curve** (e.g., steepening or flattening).

# Liability-Driven Investing (LDI): The Two-Portfolio Approach

Liability-Driven Investing (LDI) is the dominant paradigm in modern professional money management. It's a framework for managing assets that explicitly takes liabilities into account.

---

## The Central Conflict: Greed vs. Fear

Every asset owner faces a "schizophrenic" situation with two conflicting objectives:

1.  **Greed (The Quest for Performance)**: The need to generate high returns on assets. Good performance reduces the need for stakeholders to make large contributions to cover future liabilities.
2.  **Fear (The Concern for Downside)**: The worry that asset values will fall, jeopardizing the ability to meet future obligations.

In an LDI context, these objectives are viewed through the lens of the **funding ratio (Assets / Liabilities)**.
* **Greed** = The desire to *increase* the funding ratio.
* **Fear** = The concern about a sharp *decrease* in the funding ratio.

---

## The Solution: Separate the Objectives

Trying to address both performance and safety with a **single, blended portfolio** is inefficient. Such a portfolio would be schizophrenic, doing a poor job at both generating returns and managing risk.

The modern LDI approach advocates for separating these two goals into **two distinct, specialized portfolios**.

### The Skier Analogy 

Think about skiing in different conditions.
* An "all-mountain" ski tries to be good everywhere but ends up being mediocre on both groomed slopes and in deep powder.
* The better solution is to have **two specialized pairs of skis**: one narrow and stiff for speed on packed snow (performance), and one wide and soft for floating on powder (hedging/safety).

### The Two Building Blocks of LDI

1.  **The Performance-Seeking Portfolio (PSP)**
    * **Sole Focus**: To maximize risk-adjusted performance.
    * **Goal**: To efficiently harvest risk premia. The key metric is maximizing the **Sharpe Ratio**.

2.  **The Liability-Hedging Portfolio (LHP)**
    * **Sole Focus**: To hedge the risks embedded in the liabilities.
    * **Goal**: To immunize the plan's funding status against movements in risk factors like interest rates. Its value should track the value of the liabilities as closely as possible.

---

## The "Greeks" of LDI: Key Parameters

Similar to option pricing, LDI has key parameters that drive the optimal allocation between the PSP and LHP.

* **λ (Lambda) of the PSP**: This is the **Sharpe Ratio** of the performance portfolio. A higher Sharpe Ratio makes the PSP more attractive and justifies a larger allocation to it.
* **β (Beta) of the LHP**: This measures how well the LHP tracks the liabilities. The goal is a beta close to 1. The Sharpe Ratio is **not** a concern for this portfolio; its job is to hedge, not to generate excess returns.
* **σ (Sigma) of the PSP**: The volatility of the performance portfolio. All else being equal, lower volatility is preferred.
* **γ (Gamma)**: This represents the overall **risk aversion** of the investor. It's often treated as a "degree of freedom" or a policy choice that determines the overall budget allocated to the PSP versus the LHP, as institutional "risk aversion" is difficult to define.

---

## Conclusion

The core principle of LDI is that **risk and performance are distinct objectives that are best managed by two distinct building blocks**: a Performance-Seeking Portfolio and a Liability-Hedging Portfolio. The next step is to determine how to allocate funds between these two portfolios.

# Choosing the Policy Portfolio in Liability-Driven Investing (LDI)

In the LDI framework, an investor's total portfolio is split into two building blocks: a **Performance Seeking Portfolio (PSP)** for growth and a **Liability Hedging Portfolio (LHP)** for safety. The key question is: **how much should be allocated to each?**

---

## The Risk Budget Approach

While financial theory suggests this allocation should be based on an unobservable *risk aversion parameter*, the practical approach is to use a **risk budget**.

* **What is a Risk Budget?** Stakeholders (like a pension fund's Board of Trustees) define the maximum level of risk they are willing to tolerate.
* **How is it Measured?** Risk is typically measured using key metrics related to the portfolio's health relative to its liabilities, such as:
    * Volatility of the **funding ratio**.
    * Maximum drawdown of the **funding ratio**.
    * Probability of a shortfall.
* **The Goal**: The asset owner's objective is to allocate as much as possible to the **PSP** to maximize growth, without ever exceeding the pre-defined risk budget.

---

## The Fundamental Conflict: Risk Budget vs. Dollar Budget

This leads to a core dilemma in investment management, which is a trade-off between long-term goals and short-term constraints.

* **Long-Term Goal (High Return)**: To meet ambitious funding targets, especially if starting from an underfunded position, a large allocation to the PSP is needed. This requires a **large risk budget**.
* **Short-Term Constraint (Low Risk)**: Stakeholders often impose tight risk budgets, limiting the potential for short-term losses (e.g., a low maximum drawdown). This forces a small allocation to the PSP.

This creates a direct trade-off:
> If you have a **tight risk budget**, you cannot take much risk and therefore cannot expect high returns. To close a funding gap, you will need a **large dollar budget** (i.e., higher cash contributions).

---

## Example: A 40/60 vs. a 60/40 Portfolio

Let's compare a conservative portfolio (40% PSP / 60% LHP) with a more aggressive one (60% PSP / 40% LHP).

| Metric                  | 40/60 Portfolio (Conservative) | 60/40 Portfolio (Aggressive) |
| ----------------------- | ------------------------------ | ------------------------------ |
| **Median Final Funding Ratio** | 127%                           | **140%** |
| **Max Drawdown of Funding Ratio** | **16%** | 24%                            |

* **The Dilemma**: The 60/40 portfolio generates the higher long-term return needed to meet funding goals. However, its 24% maximum drawdown would likely violate a typical risk budget (e.g., a 15-20% limit).
* **The Result**: The board would be forced to choose the "safer" 40/60 portfolio, even though it is less likely to achieve the necessary long-term growth.

---

## How to Escape the Dilemma?

There are no easy answers, but there are a few real (but difficult) solutions:

1.  **Increase the Dollar Budget**: Acknowledge that the low-risk strategy won't be enough and secure **higher contributions** from stakeholders to close the funding gap.
2.  **Relax the Risk Budget**: The board makes a strategic decision to **accept more short-term risk** (e.g., a higher maximum drawdown) to allow for a more aggressive strategy that can generate higher long-term returns.
3.  **Improve the PSP**: Make the Performance Seeking Portfolio more efficient through better security selection or strategy, thereby generating a **higher return for the same amount of risk**.

"Cheating" by using overly optimistic assumptions (e.g., a high discount rate for liabilities) does not solve the underlying problem. The trade-off between risk and required capital is fundamental.

# Beyond LDI: Aligning the Performance-Seeking and Liability-Hedging Portfolios

This lecture explores how to improve upon the traditional Liability-Driven Investing (LDI) framework by considering the interaction between its two core components.

---

## A Quick Recap of Traditional LDI

The standard LDI framework separates an investor's portfolio into two distinct building blocks, which are managed in **isolation**:

1.  **Performance-Seeking Portfolio (PSP)** : Its sole focus is to **efficiently harvest risk premiums**. Success is measured by its risk-adjusted return (e.g., Sharpe ratio).
2.  **Liability-Hedging Portfolio (LHP)** : Its sole focus is to **hedge the liabilities**. Success is measured by how closely its value tracks the value of the liabilities (i.g., their correlation).

The traditional view is that if each portfolio does its job perfectly on its own, the investor's overall outcome will be optimal.

---

## Questioning the Paradigm: The Value of Interaction

This lecture challenges the "isolation" view by asking a simple question:

> If you have two PSPs with identical, high Sharpe ratios, but one of them also happens to have better liability-hedging properties, which one should you choose?

Intuition suggests you should pick the one that also helps hedge liabilities, even though that's not its primary job. This implies that the interaction between the two portfolios matters and can add value.

---

## Decomposing Investor Welfare

A formal analysis shows that an investor's welfare (measured by the final funding ratio) can be broken down into three components:

1.  **PSP Quality**: Welfare increases with the **Sharpe ratio of the PSP**. (Confirms LDI)
2.  **LHP Quality**: Welfare increases with the **correlation of the LHP to the liabilities**. (Confirms LDI)
3.  **The New Insight (The Interaction Term)**: Welfare also increases with the **correlation of the PSP to the liabilities**.

This third term is the key finding. It provides a formal justification for choosing a **"liability-friendly" PSP**, all else being equal.

---

## The Core Trade-Off: Performance vs. Hedging

The main insight is that there are strong benefits to improving the hedging characteristics of the PSP, but only if it doesn't come with a high **opportunity cost**.

Making the PSP more "liability-friendly" creates a trade-off:

* **The Positive Effect **: A more liability-friendly PSP is less risky to the overall funding ratio. This allows you to **allocate more capital** to it for the same risk budget. More dollars invested should lead to higher returns.
* **The Negative Effect **: The process of aligning the PSP with liabilities might hurt its performance, leading to a lower Sharpe ratio. This means each dollar invested now earns a **lower reward**.

The net benefit depends on which effect is stronger. The goal is to invest more dollars in the PSP *without* sacrificing too much of the reward per dollar.

---

## Conclusion and the Skiing Analogy 

While diversification (PSP) and hedging (LHP) should remain distinct goals managed by two separate portfolios, there is significant value in **aligning** them.

* **Traditional LDI**: It's best to have two specialized pairs of skis: one for groomed slopes (PSP) and one for deep powder (LHP).
* **Beyond LDI**: We still need two separate pairs of skis. A single "all-in-one" portfolio doesn't work. However, it's even better if the skis for the groomed slopes can also handle a bit of powder, and the powder skis can manage on the groomed slopes. This alignment adds value.

# Liability-Friendly Equity Portfolios

This lecture focuses on how to make the **equity** component of a Performance-Seeking Portfolio (PSP) more "liability-friendly" to improve a plan's funding ratio.

---

## The Core Question

Within the universe of stocks (e.g., the S&P 500), are there certain stocks that do a better job of hedging liabilities than others? Or are all stocks equally poor at it?

If there are differences, we can strategically select stocks to build a more effective equity portfolio.

---

## Two Approaches to Finding Liability-Friendly Stocks

### 1. The Cashflow Matching Perspective

* **The Idea**: Liabilities are a stream of cash payments. We can find stocks that generate a steady stream of cash to help meet those payments.
* **The Conclusion**: **High-dividend yield stocks** are good candidates. Their regular dividend payments can be used to pay off liabilities as they come due.

### 2. The Factor Matching Perspective

* **The Idea**: Liabilities behave like a long-duration bond portfolio. We can try to find stocks that are the most "bond-like."
* **What makes a stock "bond-like"?** Low volatility. A key feature of a fixed-income bond is the very low (or zero) volatility of its cashflows (coupons). Low-volatility stocks are the closest approximation to this feature within the equity universe.
* **The Conclusion**: **Low-volatility (low-vol) stocks** are good candidates.

---

## Empirical Evidence: Do These Stocks Actually Work?

A comparison of the tracking error of different equity portfolios against a 10-year duration liability proxy shows significant differences:

* **S&P 500 Index**: Tracking Error = **18.8%**
* **High-Volatility Stocks**: Tracking Error = **28.0%**
* **Low-Volatility Stocks**: Tracking Error = **14.0%**

**Finding**: Yes, there is a large difference. Low-vol stocks have a tracking error that is **half** that of high-vol stocks, confirming they are significantly more liability-friendly.

> **Note**: This does not mean low-vol stocks are a substitute for bonds in a Liability Hedging Portfolio (LHP). It means that when building the *equity* portion of the portfolio, tilting towards low-vol stocks is a better choice from a liability perspective.

---

## The Benefit in Practice: Higher Allocation, Better Outcome

Let's see how this improves a portfolio.

* **Baseline Scenario**: A traditional `40/60` portfolio (40% S&P 500, 60% bonds) has a funding ratio volatility of **7.5%**. Let's assume this is the maximum allowable risk budget.

* **Step 1: Substitute the Equity**: Replace the S&P 500 with a low-vol equity portfolio.
    * In a `40/60` mix, the funding ratio volatility **drops from 7.5% to 5.84%**. The portfolio is now much safer relative to the liabilities.

* **Step 2: Increase the Allocation**: Since we now have "room" in our risk budget, we can increase our allocation to this more liability-friendly equity portfolio.
    * We can increase the equity allocation from **40% up to 51%** before the funding ratio volatility gets back up to our 7.5% risk limit.

### The Final Outcome

By selecting a more liability-friendly equity portfolio, we can allocate a larger portion of our assets to it for the **same risk budget**. This leads to superior performance for two reasons:
1.  **More Dollars Allocated**: We simply have more money working in the equity market.
2.  **No Performance Trade-off**: Historically, liability-fr
iendly strategies like low-vol and high-dividend have performed as well as, or even better than, the broad market.

## Chapter 1: Bond Pricing Fundamentals 

### 1.1 The Price-Yield Relationship

The price of a bond is the **present value of all its future cash flows** (coupon payments and the final principal repayment), discounted at the current market yield, also known as the Yield to Maturity (YTM). The relationship between a bond's fixed coupon rate and the market yield determines whether it trades at a premium, par, or discount.

* **Premium Bond (Price > Par):** Occurs when the **Coupon Rate > Market Yield**. Investors are willing to pay more for a bond that offers higher coupon payments than what the current market offers.
    * *Example:* A 5-year bond with a 6% coupon in a 5% yield environment will be the **most expensive**, trading at a premium (e.g., ~$1044).

* **Par Bond (Price = Par):** Occurs when the **Coupon Rate = Market Yield**. The bond's promised payments are exactly in line with market expectations, so it trades at its face value (typically $1000).
    * *Example:* A 15-year bond with a 5% coupon in a 5% yield environment will trade at par ($1000).

* **Discount Bond (Price < Par):** Occurs when the **Coupon Rate < Market Yield**. Since the bond pays less than the prevailing market rate, its price must be lower to attract investors.

### 1.2 The Special Case: Zero-Coupon Bonds

A zero-coupon bond makes no periodic interest payments. Its only cash flow is the face value paid at maturity. Therefore, its price is simply the face value discounted back to the present.

The formula is:
$$ \text{Price} = \frac{\text{Face Value}}{(1 + \text{Yield})^N} $$
Where N is the number of years to maturity.

* *Example:* A 10-year zero-coupon bond in a 5% yield environment is priced by discounting its $1000 face value back 10 years. This results in a price of approximately **$614**, making it the **least expensive** of the options.

## Chapter 2: Understanding Bond Duration 

### 2.1 What is Macaulay Duration?

**Macaulay Duration** is a measure of a bond's interest rate sensitivity. It represents the weighted-average time (in years) until an investor receives the bond's cash flows. A higher duration implies that a bond's price will change more significantly for a given change in interest rates. 

Duration is influenced by three main factors:

1.  **Maturity:** Longer maturity generally leads to higher duration.
2.  **Coupon Rate:** A **higher coupon rate leads to a lower duration**. This is because the investor receives more cash back sooner, lowering the weighted-average time of the cash flows.
3.  **Yield:** A higher yield leads to a lower duration, as future cash flows are discounted more heavily.

### 2.2 Comparing Durations

Let's compare three distinct bonds:
* **B1 (15-year, 5% coupon):** Has a long maturity, giving it a high duration of **~10.73 years**.
* **B2 (5-year, 6% coupon):** Has both a short maturity and a high coupon rate, resulting in the **lowest duration** of **~4.37 years**.
* **B3 (10-year, 0% coupon):** For any zero-coupon bond, the **Macaulay Duration is always equal to its maturity**. Therefore, its duration is exactly **10 years**.

Based on this, B1 has the highest duration, and B2 has the lowest.

## Chapter 3: Liability Driven Investing (LDI) and Duration Matching

### 3.1 Calculating the Duration of Liabilities

Just like a portfolio of bonds, a series of future liabilities (e.g., pension payments) also has a duration. The **Duration of Liabilities** is the weighted-average timing of these future cash outflows, where the weights are determined by the present value of each liability. This single number represents the interest rate sensitivity of the total liability obligation.

* *Example:* For liabilities of $100k, $200k, and $300k due in 3, 5, and 10 years respectively, the present-value-weighted timing results in a liability duration of **6.75 years**.

### 3.2 The Strategy of Duration Matching

**Duration Matching** is an immunization strategy used to manage interest rate risk. The goal is to build a portfolio of assets (bonds) whose duration is equal to the duration of the liabilities. When matched, any change in interest rates should affect the value of the assets and liabilities equally, protecting the net value of the portfolio.

To create a two-bond portfolio with a target duration, we solve a system of two equations, where $w_A$ and $w_B$ are the weights of Bond A and Bond B:

1.  $w_A + w_B = 1$
2.  $w_A \cdot D_A + w_B \cdot D_B = D_{\text{Target}}$

### 3.3 The "Bracketing" Principle

A critical rule for building a duration-matched portfolio is that the target duration **must lie between the durations of the bonds being combined**. You cannot create a portfolio with a duration that is shorter than your shortest-duration asset or longer than your longest-duration asset.

* **Feasible Combination:** To match a liability duration of **6.75 years**, you can combine Bond B2 (Duration = 4.37) and Bond B3 (Duration = 10.0). The target lies between them. Solving the equations gives a weight of **57.73%** for B2.
* **Infeasible Combination:** You **cannot** use Bond B1 (Duration = 10.73) and Bond B3 (Duration = 10.0) to match the 6.75-year target. Since both bonds have durations longer than the target, any combination of them will also have a duration longer than 6.75.

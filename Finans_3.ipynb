# Extensions of Constant Proportion Portfolio Insurance (CPPI)

This lecture discusses two useful extensions to the basic CPPI strategy:
1.  Protecting a **maximum drawdown floor**.
2.  Protecting a portfolio with both a **floor and a cap**.

---

## 1. Protecting a Maximum Drawdown Floor

The goal of this strategy is to ensure that the portfolio's value never falls more than a predefined percentage from its highest point ever reached. 


### Key Concepts

* **Asset Value Process ($V_t$)**: The value of your portfolio at time $t$.
* **Running Max Process ($M_t$)**: Keeps track of the maximum value the portfolio has reached from time 0 to time $t$.
    * Mathematically: $M_t = \max_{0 \le s \le t} V_s$
* **Maximum Drawdown Constraint**: The strategy imposes a constraint that the current value $V_t$ must always be greater than a certain percentage ($\alpha$) of the running maximum $M_t$.
    * Constraint: $V_t > \alpha \cdot M_t$
    * This ensures the maximum drawdown is limited to $(1 - \alpha)$.

### Example: 20% Maximum Drawdown

* If you want to limit the maximum drawdown to **20%**, you set $(1 - \alpha) = 0.20$, which means $\alpha = 0.80$.
* The constraint becomes: $V_t > 0.80 \cdot M_t$.
* The **floor** is now dynamic and is defined as $F_t = 0.80 \cdot M_t$.

### Characteristics of the Maximum Drawdown Floor

* The floor ($F_t = \alpha \cdot M_t$) is a **non-decreasing process**.
* When the portfolio value $V_t$ increases and hits a new high, the running maximum $M_t$ increases, and therefore the floor $F_t$ also increases.
* When the portfolio value $V_t$ decreases, the running maximum $M_t$ remains at its last peak, so the floor $F_t$ stays flat. It **never decreases**.

### Implementation

To implement this strategy, you use the standard CPPI formula but with this dynamically defined floor:
* **Exposure to Risky Asset** = $m \cdot (\text{Asset Value} - \text{Floor})$
* **Formula**: $E_t = m \cdot (V_t - F_t) = m \cdot (V_t - \alpha \cdot M_t)$
    * Where $m$ is the multiplier.

---

## 2. CPPI with a Floor and a Cap

This strategy is designed for investors who have a minimum amount of wealth they want to protect (a floor) but also a maximum target wealth (a cap) beyond which they have little desire to go. 

### Rationale

* Protecting downside (the floor) has an **opportunity cost** (giving up potential gains).
* By also giving up on upside gains beyond a certain cap, the investor can generate an **opportunity gain**.
* This opportunity gain can offset some or all of the opportunity cost of the downside protection.
* This is useful for investors with a specific target or goal in mind.

### How it Works

The strategy uses a **threshold level ($T_t$)** that is set between the floor ($F_t$) and the cap ($C_t$). The allocation to risky assets changes depending on whether the portfolio value is above or below this threshold.

1.  **Below the Threshold ($F_t \le V_t < T_t$)**:
    * The focus is on protecting the floor.
    * The strategy behaves like standard CPPI.
    * **Allocation**: $E_t = m \cdot (V_t - F_t)$

2.  **Above the Threshold ($T_t < V_t \le C_t$)**:
    * The focus shifts from protecting the floor to avoiding overshooting the cap.
    * The strategy starts reducing risk as the portfolio value gets closer to the cap.
    * **Allocation**: $E_t = m \cdot (C_t - V_t)$

### Determining the Threshold ($T_t$)

The threshold is chosen to ensure a smooth transition in allocation. This is achieved through a **"smooth-pasting condition"**, which requires that the allocation formula is the same whether you approach the threshold from below or from above.

* At the exact moment $V_t = T_t$, the two allocation rules must be equal:
    $$m \cdot (T_t - F_t) = m \cdot (C_t - T_t)$$
* Solving for $T_t$:
    $$T_t - F_t = C_t - T_t$$
    $$2T_t = C_t + F_t$$
    $$T_t = \frac{F_t + C_t}{2}$$
* The threshold is simply the **average (midpoint)** of the floor and the cap.

# Generating Asset Return Scenarios with a Random Walk Model

This is a foundational method for generating scenarios for asset returns using a **random walk model**. This is a core technique for performing Monte-Carlo simulations in finance. We'll look at a simple case with two assets: one risky and one risk-free. 

---

## The Two-Asset Model

### 1. The Risky Asset (e.g., a Stock Index)

The value of the risky asset at time $t$ is denoted by $S_t$. Its return over an infinitesimally small time period, $dt$, is modeled by a stochastic differential equation.

The model for the risky asset's return is:
$$ \frac{dS_t}{S_t} = \mu dt + \sigma dW_t $$

This equation states that the percentage return ($\frac{dS_t}{S_t}$) has two main parts:

* **A Deterministic "Trend" Component ($\mu dt$)**: This represents the expected return of the asset.
    * $\mu$ is the **annualized expected return**.
    * This component is often broken down further to better understand the drivers of return:
        $$ \mu = r + \lambda \sigma $$
        * $r$: The **risk-free rate**.
        * $\sigma$: The **volatility** of the asset (a measure of its risk).
        * $\lambda$: The **Sharpe Ratio** (the price of risk, or return per unit of risk).
        * The term $\lambda \sigma$ is the **risk premium**â€”the excess return you expect for taking on risk.
    * **Example**: If $r = 2\%$, $\sigma = 16\%$, and $\lambda = 0.5$, then the expected return is $\mu = 2\% + (0.5 \times 16\%) = 10\%$.

* **A Stochastic "Noise" Component ($\sigma dW_t$)**: This represents the random, unpredictable up-and-down movements of the asset around its trend.
    * $\sigma$ is the volatility, which scales the magnitude of the random shocks.
    * $dW_t$ is the increment of a **Brownian Motion** process, which is our model for a pure random walk.

### 2. The Risk-Free Asset (e.g., a T-Bill)

The value of the risk-free asset at time $t$ is denoted by $B_t$. Its return is entirely predictable.

The model for the risk-free asset's return is:
$$ \frac{dB_t}{B_t} = r dt $$

Notice there is no stochastic component ($\sigma dW_t$). By definition, its return over any period is known with certainty.

---

## Understanding Brownian Motion ($W_t$)

Brownian motion is the core building block for the randomness in our model. 

### History and Concept

* It's a mathematical model for a **pure random walk**â€”a path made of random steps.
* It was first described by French mathematician **Louis Bachelier** in 1900 for his Ph.D. thesis on option pricing.
* It was independently rediscovered by **Albert Einstein** in 1905 to describe the random motion of particles suspended in a fluid.

### Discrete-Time Formulation

To make it more intuitive, we can think about the model in discrete time steps (e.g., from one day to the next, $\Delta t$). The return from $t$ to $t+\Delta t$ is:
$$ \frac{S_{t+\Delta t} - S_t}{S_t} \approx \mu \Delta t + \sigma \sqrt{\Delta t} Z_t $$
where $Z_t$ is a random variable drawn from a standard normal distribution (mean=0, variance=1).

From this, we can see two key properties for returns over an interval $\Delta t$:
1.  **Expected Return**: $E[\text{Return}] = \mu \Delta t$
2.  **Variance of Return**: $Var[\text{Return}] = \sigma^2 \Delta t$

This scaling property is why volatility is often quoted in annualized terms ($\sigma$) and then scaled by the square root of time for shorter periods.

### Key Properties of Brownian Motion

* **Independent Increments**: The movement in the next time step is completely independent of all past movements. Whether the asset went up or down yesterday has no bearing on where it will go today.
* **No Serial Correlation**: The model assumes there is no **momentum** (a trend is likely to continue) or **mean-reversion** (an asset is likely to revert to its average). It is a pure, memoryless random walk.

---

## Conclusion

The Brownian motion model provides a simple yet powerful framework for simulating asset returns.

* It serves as a fundamental **building block** for generating random scenarios for risky assets like stocks.
* For risk-free assets (over short horizons), no randomness is needed; their returns are deterministic.
* While this is a foundational model, it's a "fair approximation." More advanced models, which will be discussed later, can be used to add more realistic features like changing volatility or correlations.

# Monte-Carlo Simulation with Time-Varying Parameters

This lecture builds upon the basic random walk model by introducing a more realistic feature: **time-varying parameters**. The goal is to generate more realistic and robust financial scenarios by acknowledging that market conditions are not static. 

---

## From Constant to Dynamic Models

### Recap: The Simple Model

Last time, we introduced a simple model for a risky asset's return:
$$ \frac{dS_t}{S_t} = (r + \lambda \sigma) dt + \sigma dW_t $$
A key limitation of this model is that it assumes the risk-free rate ($r$), Sharpe ratio ($\lambda$), and volatility ($\sigma$) are all **constant**.

### The Reality: Parameters Change Over Time

In the real world, these parameters are constantly changing:
* **Interest Rates ($r$)**: Central bank policies and economic conditions cause interest rates to fluctuate.
* **Volatility ($\sigma$)**: Markets experience periods of high and low volatility. Volatility is not constant.
* **Sharpe Ratio / Risk Premia ($\lambda$)**: The compensation investors demand for taking on risk also changes based on market sentiment and economic outlook.

To make our simulations more realistic, we must treat these parameters as dynamic processes themselves. We replace the constants with time-varying quantities:
* $r \rightarrow r_t$
* $\sigma \rightarrow \sigma_t$ (or we can model the variance, $V_t = \sigma_t^2$)
* $\lambda \rightarrow \lambda_t$

Our asset return model now becomes:
$$ \frac{dS_t}{S_t} = (r_t + \lambda_t \sigma_t) dt + \sigma_t dW_t $$

---

## Modeling the Parameters: Mean-Reverting Processes

If the parameters are no longer constant, how do we model their evolution? A common and intuitive choice is to use **mean-reverting processes**.

### What is Mean Reversion? 

A mean-reverting process is one that tends to be pulled back towards a long-term average or "mean." Unlike a pure random walk which can wander off to infinity, these variables tend to stay within a typical range.

Think of it like a thermostat in a room. If the temperature gets too high, the AC kicks in to cool it down. If it gets too low, the heater turns on to warm it up. The temperature is always pulled back towards the target setting. Financial parameters like interest rates and volatility often behave this way.

### 1. Modeling the Risk-Free Rate ($r_t$)

We can model the interest rate using a process that pulls it toward a long-term average, $b$. A common model (the Vasicek model) is:
$$ dr_t = a(b - r_t)dt + \sigma_r dW_t^r $$
* $b$: The **long-term mean** interest rate. This is the level that $r_t$ is pulled towards.
* $a$: The **speed of mean reversion**. A higher `a` means the rate snaps back to its mean more quickly.
* $(b - r_t)$: The "mean-reverting drift."
    * If $r_t < b$ (rate is below average), the term is positive, pushing the rate **up**.
    * If $r_t > b$ (rate is above average), the term is negative, pulling the rate **down**.
* $\sigma_r dW_t^r$: A random shock term, with its own volatility ($\sigma_r$) and Brownian motion ($dW_t^r$).

### 2. Modeling Variance ($V_t = \sigma_t^2$)

Variance can also be modeled as a mean-reverting process. However, there's a critical constraint: **variance can never be negative!**

To ensure this, a special type of mean-reverting process called a **square-root process** is often used (e.g., the CIR or Heston model):
$$ dV_t = \kappa(\theta - V_t)dt + \xi \sqrt{V_t} dW_t^V $$
* $\theta$: The long-term mean variance.
* $\kappa$: The speed of mean reversion.
* The key feature is the $\sqrt{V_t}$ term multiplying the random shock. As variance ($V_t$) gets close to zero, this term also gets very small, effectively "turning off" the random shocks and preventing the variance from dropping below zero.

---

## Conclusion

* To generate realistic financial scenarios, it's crucial to move beyond simple models with constant parameters.
* By modeling key drivers like **interest rates, volatility, and risk premia** as their own stochastic processes, we can capture a much wider and more plausible range of future outcomes.
* **Mean-reverting processes** are a powerful and intuitive tool for modeling these parameters, as they reflect the real-world tendency of these variables to fluctuate around long-term averages.

# Understanding CPPI Parameters: An Interactive Look

This lecture serves as a bridge between theory and practice, providing an intuitive feel for the parameters that drive asset simulations and CPPI strategies. We'll explore how changing these parameters affects the potential outcomes of an investment. 

---

## Part 1: The Building Blocks - Geometric Brownian Motion (GBM)

Before applying CPPI, we must understand the engine that generates our scenarios: the Geometric Brownian Motion model. It's governed by two key parameters: **drift ($\mu$)** and **volatility ($\sigma$)**.

### The Drift Parameter ($\mu$)

The drift, $\mu$, dictates the **average trend** of the asset price over time.

* **When $\mu = 0$**: The simulation is a pure **random walk with no drift**. The price paths wander randomly, but on average, they end up right where they started. There is no built-in upward or downward pull.
    
* **When $\mu > 0$** (e.g., $\mu = 7\%$): The paths now have a clear **upward trend**. While individual paths are still random, the vast majority end up above the starting point. The drift acts like a gentle, persistent wind pushing the average outcome higher.
    

### The Volatility Parameter ($\sigma$)

The volatility, $\sigma$, controls the **dispersion** or "fan" of possible outcomes. It measures how wildly a single path can deviate from its trend.

* **When $\sigma$ is low**: The random paths are tightly clustered around the drift. There is very little uncertainty, and the final outcomes are all very similar. The journey is almost a straight line.
* **When $\sigma$ is high** (e.g., $\sigma = 30\%$): The dispersion of outcomes explodes. The "fan" of possibilities becomes extremely wide, leading to some outrageously high potential returns but also significant potential losses. A higher sigma means a wider range of possible futures.
    

---

## Part 2: Introducing CPPI - The Art of "Outcome Shaping"

Now, let's see what happens when we run a CPPI algorithm on top of these simulated scenarios. CPPI fundamentally alters the final distribution of wealth.

### The Problem: A Typical Return Distribution

Without any protection, the final wealth from a stock-only investment typically follows a **log-normal distribution**. This distribution has a long right tail (potential for high gains) but also a meaningful left tail, representing the risk of significant losses.


### The Solution: CPPI and the Floor

When we introduce a CPPI strategy with a **floor** (e.g., a floor at 70% of initial wealth), the magic happens.

* **The Mechanism**: The CPPI algorithm continuously monitors the portfolio's **cushion** (Account Value - Present Value of the Floor) and uses a **multiplier ($m$)** to decide how much to allocate to the risky asset.
* **The Result**: The CPPI strategy effectively **cuts off the left tail** of the distribution. The outcomes that would have previously ended below the floor are now protected. The algorithm de-risks the portfolio as it approaches the floor, preventing breaches.

This transformation is called **"outcome shaping"**. We are actively reshaping the probability distribution of our final wealth to align with our goalsâ€”specifically, eliminating the worst-case scenarios.


### A Note on Floor Violations

In the lecture's initial simulation (with modest volatility and parameters), there were **zero floor violations** across 1,000 scenarios. This is a crucial insight:

> CPPI can be highly effective even with periodic, not continuous, trading (e.g., quarterly checks in the simulation).

However, this isn't a guarantee. **Floor violations** become more likely when:
* Market volatility ($\sigma$) is very high.
* The CPPI multiplier ($m$) is aggressive.
* Trading is infrequent, allowing for large gaps where the portfolio can't be rebalanced.

Exploring the trade-offs between these parameters is key to designing a robust CPPI strategy.

# CPPI: Challenges in Volatile Markets and Adjusting Parameters

This lecture explores the performance of the Constant Proportion Portfolio Insurance (CPPI) strategy under stressful market conditions and examines how to adjust its parameters to control risk.

---

## The Problem: High Volatility Leads to Floor Violations

We start with a CPPI model that works perfectly in a calm market (e.g., 15% volatility) with quarterly rebalancing, showing **no floor violations**. However, this ideal scenario may not hold in the real world.

### The Impact of Increasing Volatility

When we increase the volatility of the risky asset (the PSP or "Principal Strategic Portfolio"), the strategy's protection begins to fail.

* **15% -> 20% Volatility**: Floor violations begin to appear (2 out of 1000 cases).
* **25% Volatility**: The number of violations increases significantly.
* **35% Volatility**: The failure rate becomes critical, with violations in **~20%** of all scenarios.

### Why Do Violations Happen? The "Gap Risk"

The core issue is the combination of **high volatility** and **discrete rebalancing periods** (e.g., quarterly).

A **floor violation** occurs when the market drops so sharply *between* rebalancing dates that the portfolio's value falls through the floor before we have a chance to de-risk by selling the risky asset. By the next rebalancing date, it's already too late. This is often called **gap risk**. 



---

## Solutions to Mitigate Floor Violations

There are two primary ways to fix this problem.

### 1. Increase Rebalancing Frequency

The most straightforward solution is to react more quickly to market changes.

* **The Idea**: Don't wait a full quarter. Rebalance more often.
* **The Result**: By increasing the rebalancing frequency from quarterly (4 times/year) to **weekly (52 times/year)**, the violations are almost completely **eliminated**, even in a high-volatility (35%) environment.
* **Takeaway**: More frequent rebalancing significantly reduces gap risk.

### 2. Adjust the Multiplier ($m$)

What if we must stick to quarterly rebalancing? We can still control risk by adjusting the CPPI parameters, specifically the **multiplier ($m$)**.

The multiplier controls the **aggressiveness** of the strategy. A higher $m$ means more leverage and a larger allocation to the risky asset for a given cushion.

#### In a High Volatility Market (35% Volatility)

To reduce violations, we must be less aggressive. We can **decrease the multiplier ($m$)**.

* $m=3.0 \rightarrow$ ~20% Violations
* $m=2.5 \rightarrow$ ~10% Violations
* $m=2.0 \rightarrow$ ~3% Violations
* $m=1.5 \rightarrow$ Almost 0 Violations
* $m=1.0 \rightarrow$ **0 Violations**

#### The Special Case: $m = 1$

An $m=1$ is the most conservative setting. It means the amount invested in the risky asset is **exactly equal to the cushion**. The floor value itself is held in the safe asset. To violate the floor, the entire cushion would have to be wiped out in a single period, which is highly unlikely.

---

## The Trade-Off: Performance in Calm Markets

Using a low multiplier is safe, but it comes at a cost, especially in calm markets.

### The Downside of Being Too Conservative

In a **low-volatility market (e.g., 10%)**, a low multiplier like $m=1$ is overly cautious. It limits investment in the risky asset, causing the portfolio to miss out on potential upside from the market's positive drift.

As we **increase $m$ in a calm market**:

* The average final wealth **steadily increases**.
* Crucially, **no new violations appear**. The market is not volatile enough to cause large drops that would pierce the floor, even with higher leverage (e.g., $m=5$).

---

## Conclusion: The Dynamic Multiplier

The key takeaway is the direct relationship between the market environment and the optimal CPPI parameters.

* **High Volatility Market **: Use a **small $m$** to prioritize safety and avoid floor violations.
* **Low Volatility Market **: Use a **large $m$** to increase market participation and enhance returns.

This suggests that a fixed multiplier is suboptimal. In practice, a **dynamic multiplier** is often used, where $m$ is adjusted over time to reflect current or expected market conditions.

# Theory and Analysis of Constant Proportion Portfolio Insurance (CPPI)

---

## Chapter 1: Fundamentals of CPPI Strategy and Analysis

### 1.1 The CPPI Mechanism

Constant Proportion Portfolio Insurance (CPPI) is a dynamic asset allocation strategy that allows an investor to maintain exposure to a risky asset while providing a level of protection against downside risk. The core of the strategy is governed by three key components:

* **Assets:** The total current value of the portfolio.
* **Floor:** The minimum desired value of the portfolio, below which the investor does not want the assets to fall.
* **Cushion:** The difference between the current assets and the floor (`Cushion = Assets - Floor`). This represents the amount of capital that can be risked.

The allocation to the risky asset is determined by a simple formula, where **m** is the **multiplier**, a constant that determines the strategy's aggressiveness:

$$ \text{Exposure to Risky Asset} = m \times \text{Cushion} $$

The remainder of the portfolio is invested in a safe, risk-free asset.

### 1.2 Using Monte Carlo Simulation for Analysis

To understand the potential outcomes of a CPPI strategy, we use **Monte Carlo Simulation**. This method involves running thousands of randomized scenarios for the price path of the risky asset based on its expected return ($\mu$) and volatility ($\sigma$).

A key insight from this analysis is that as the **number of scenarios increases**, the simulation provides a more complete picture of the probability distribution of outcomes. This means it becomes more likely to capture extreme tail events. Consequently, the **difference between the best-case and worst-case terminal wealth will increase**, not because the strategy is riskier, but because the analysis is more thoroughly exploring the full range of possibilities. 

---

## Chapter 2: Impact of Key Parameters on CPPI Performance

The behavior and final outcome of a CPPI strategy are highly sensitive to the parameters chosen by the investor and the prevailing market conditions.

### 2.1 The Protective Floor

The **floor** is the foundational element of the strategy's protective mechanism. As an investor **increases the floor**, the cushion (`Assets - Floor`) decreases. This mechanically leads to a smaller allocation to the risky asset, making the overall strategy more conservative. As a result, a higher floor **increases the portfolio's value in the worst-case scenario** by limiting the potential for significant losses.

### 2.2 Market Parameters: Return and Volatility

The characteristics of the risky asset directly impact the portfolio's performance:

* **Expected Return ($\mu$):** This parameter represents the average expected growth rate or drift of the risky asset. All else being equal, a **higher $\mu$ will lead to a higher expected terminal wealth**. Since the portfolio has exposure to this asset, its positive drift naturally lifts the entire portfolio's expected performance over time.
* **Volatility ($\sigma$):** This parameter measures the magnitude of the risky asset's price swings. Higher volatility implies greater uncertainty and larger potential drawdowns. An **increase in $\sigma$ makes a sudden, large price drop more likely**, thereby increasing the probability of a negative final return or a floor violation.

### 2.3 The Multiplier and Rebalancing Frequency

The multiplier (**m**) and the rebalancing frequency are the primary levers for controlling the strategy's risk profile:

* **The Multiplier (m):** A multiplier greater than 1 ($m > 1$) introduces leverage, making the strategy more aggressive. It amplifies both gains and losses. A higher 'm' significantly increases the risk of the portfolio's value falling through the floor.
* **Rebalancing Frequency:** The strategy must be periodically rebalanced to adjust the risky asset allocation according to the formula. Infrequent rebalancing creates **gap risk**â€”the danger that a sharp market downturn between rebalancing dates can cause the portfolio's value to "gap down" through the floor before any defensive action can be taken.

Therefore, the combination of a **high multiplier and low rebalancing frequency** is particularly risky, as it pairs an aggressive allocation with a slow reaction time, substantially increasing the likelihood of floor violations.

---

## Chapter 3: Measuring Risk in CPPI

### 3.1 Floor Violations

A **floor violation** occurs when the portfolio's value drops below the pre-determined floor. The probability of a violation is a primary measure of the strategy's risk. Factors that increase the likelihood of a floor violation include:
* A high multiplier ($m$).
* High asset volatility ($\sigma$).
* Infrequent rebalancing.

### 3.2 Expected Shortfall

**Expected Shortfall** is a more advanced risk metric that answers the question: "If a floor violation happens, what is the average loss we can expect?" It measures the severity of the failure.

There is a direct relationship between the probability of a violation and its potential severity. The parameter changes that make floor violations more likely (e.g., increasing 'm', decreasing rebalancing frequency) also tend to make the violations more severe when they occur. A more aggressive and slower-to-react strategy not only fails more often but also fails more spectacularly. Therefore, **factors that increase floor violations also tend to increase the Expected Shortfall**.

## Chapter 4: CPPI for Principal Protection

### 4.1 Conditions for Principal Protection

A common application of CPPI is for **principal protection**, where the floor is set equal to the initial investment. The goal is to ensure the investor at least gets their initial capital back.

This guarantee, however, is critically dependent on the **risk-free rate being non-negative ($\geq 0$)**. The strategy's ultimate safety net is its ability to shift all capital into the safe asset once the portfolio value hits the floor. If the risk-free rate is negative, this "safe" asset actually erodes in value over time, making it impossible to guarantee the principal will be preserved.

### 4.2 Achieving Zero Expected Shortfall

In theory, to achieve zero floor violations and thus zero expected shortfall, the strategy's inherent leverage risk must be eliminated. This can be achieved by setting the multiplier **m = 1**. With $m=1$, the exposure to the risky asset is equal to the cushion. In this case, a catastrophic 100% loss in the risky asset would only wipe out the cushion, leaving the portfolio value exactly at the floor. While discrete rebalancing still introduces minor gap risk, setting $m=1$ is the fundamental condition required to remove the risk of overshooting the floor.
